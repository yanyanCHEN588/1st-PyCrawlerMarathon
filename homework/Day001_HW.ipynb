```python
from urllib.request import urlretrieve
import os

#------------------------------------下載檔案並存檔取名-------------------------------------------------
try:
    os.makedirs( './Data', exist_ok=True )  #exist_ok=True 是可以創建一個不存在的目錄 若已存在即跳過
    
    urlretrieve ("https://www.w3.org/TR/PNG/iso_8859-1.txt", "./1.txt") #只是示範
    urlretrieve ("https://www.w3.org/TR/PNG/iso_8859-1.txt", "./Data/Homework.txt") #示範存在創立之Data資料夾
except:
    print('發生錯誤！' )

#---------------------------- 檢查 Data 資料夾是否有 Homework.txt 檔名之檔案-------------------------    
#**人工判讀

# 打開文件（預設位置會是跟 .ipynb 程式相同的目錄）
dirs = os.listdir( './' )         
# 输出所有文件和文件夹 顯示看看阿        
for file in dirs:
    print(file)
    
print('\n')

dirs = os.listdir( './Data' )         
# 输出所有文件和文件夹 顯示看看阿        
for file in dirs:
    print(file)
print('\n')

#**編碼判讀
files = []
files = os.listdir( './Data' ) 
if 'Homework.txt' in files:
    print('[O]  Data 資料夾  有  ^^ Homework.txt 檔名之檔案')
else:
    print('[X]  Data 資料夾 沒有:(( Homework.txt 檔名之檔案')
    
print('\n') 
#-----------------------------------將「Hello World」字串覆寫到 Homework.txt 檔案  真的是直接覆蓋餒 之前裡面大筆資料都掰掰了----------------

with open("./Data/Homework.txt", "w") as fh:
    f = fh.write("Hello World")
    print(f)
    print('\n')
    
with open("./Data/Homework.txt", "r") as fh:
    f = fh.read()
    print(f)

# 檢查 Homework.txt 檔案字數是否符合 Hello World 字數

if len('Hello World') == len(f):
    print('[O] 檢查 Homework.txt 檔案字數是否符合 Hello World 字數')
else:
    print('[X] 檢查 Homework.txt 檔案字數是否符合 Hello World 字數')   
```
